## Data Cleaning 

```python
# Checking for null values
for column in df.columns:
    print(column, ", n(missing values) = ", df[column].isnull().sum()*100/len(df), ", n unique values = ", df[column].nunique())
    print('')
```

```python
df['BP'].value_counts()
```

```python
df['BP'].replace(['HIGH', 'NORMAL', 'LOW'], [2, 1, 0], inplace=True)
```

```python
df['Cholesterol'].replace(['HIGH', 'NORMAL'], [1, 0], inplace=True)
```

```python
df.head()
```

### Converting Pandas DataFrame to Numpy arrays

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[0:5]
```

```python
y = df[['Drug']].values
y[0:5]
```

## Decision Tree

## Train-Test-Split

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
for i in range(0, 200): #rows are 200
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=i)
    
    #creating Decision Tree classifier
    drugtree = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=i)
    drugtree.fit(X_train, y_train)
    
    #determining the score
    train_score = drugtree.score(X_train, y_train)
    test_score = drugtree.score(X_test, y_test)
    
    print("Test: {}, Train: {} and Random State: {}".format(train_score, test_score, i))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)
```

```python
print("Train sets: ", X_train.shape, y_train.shape)
print("Test sets: ", X_test.shape, y_test.shape)
```

```python
drugtree = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=2)
drugtree
```

```python
drugtree.fit(X_train, y_train)
```

```python
yhat = drugtree.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
print("Decision Tree's accuracy = ", accuracy_score(y_test, yhat))
```
